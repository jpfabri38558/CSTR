```python
from dataclasses import dataclass

@dataclass
class parametersCSTR:
  """
  Classe de parâmetros do CSTR.
  k - Constante de velocidade (1/min)
  F - Vazão de entrada (L/min)
  V - Volume do reator (L)
  """
  k: float
  F: float
  V: float

def modelCSTR(CAin, parameters):
  """
  Calcula o valor da concentração de saída do CSTR.
  CAin - concentração de entrada do reator
  parameters - parâmetros do reator (k, F, V)
  """
  k = parameters.k
  F = parameters.F
  V = parameters.V
  return (F*CAin)/(F + k*V)

def simCSTR(N = 3):
  """
  Simula uma série de N CSTRs, retornando as concentrações
  de saída de cada reator e a conversão acumulada.
  N - número de reatores
  CA - lista com as concentrações de saída de cada reator
  X - lista com a conversão acumulada de cada reator
  parameters - parâmetros do reator (k, F, V)
  """
  CA = [CA0]
  X = []
  parameters = parametersCSTR(k, F, V)

  for i in range(N):
    CA.append(modelCSTR(CA[i], parameters))
    X.append((CA[0] - CA[i+1])/CA[0])

  return CA, X
```
